# Quiz 1
สร้าง DataFrame ที่เก็บยอดขายของบริษัท A,B,C,D ของปี 2014,2015,2016,2017 โดยชื่อบริษัทเป็น index และปีเป็น column. สร้างคำสั่ง numpy สุ่มเลขแบบ float ให้เป็นรายได้ของบริษัทที่อยู่ในช่วง 50 – 500 (หน่วยล้านบาท) 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
name = ['A' ,'A','A','A', 
        'B' ,'B','B','B',
        'C' ,'C','C','C', 
        'D','D','D','D']
year = ['2014','2015','2016','2017',
       '2014','2015','2016','2017',
       '2014','2015','2016','2017',
       '2014','2015','2016','2017']
sales = np.random.uniform(50,500,16)
datas = list(zip(name,year,sales))
cols = ['Name', 'Year' , 'Sales(Hundred million Baht)']
df = pd.DataFrame(datas, columns = cols)
df.set_index(['Name'] , inplace=True)
df
```

# Quiz 2
ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Bar plot ให้แกน x เป็น ชื่อบริษัท ยอดขายเฉลี่ยเป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
sns.barplot(x=df.index,y='Sales(Hundred million Baht)',data=df)
```

```python
sns.barplot(y=df.index,x='Sales(Hundred million Baht)',data=df)
```
