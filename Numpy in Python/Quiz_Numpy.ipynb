# Quiz 5
5. นำข้อ 3 และ 4 มาเปรียบเวลากัน โดยใช้ฟังก์ชั่น time.time( ) เพื่อจับเวลา แสดงผลลัพธ์ในรูปแบบมิลลิวินาที

```python
import time
#Solution Numpy
time1 = time.time()
num = np.random.uniform(10,99,20)
num**2
time2 = time.time()
time_array = time2 - time1
print(time_array)
```

```python
# Solution List Comprehension
time1 = time.time()
x =np.random.uniform(10,99,20)
num2 = list(x)
num3 = [i**2 for i in num2]
time2 = time.time()
time_list = time2 - time1
print(time_list)
```

20. เขียนโปรแกรมรับค่าจากผู้ใช้ให้ป้อนเลข 9 ตัว แต่ละตัวต้องไม่ซ้ำกันและมีค่าอยู่ในช่วง 10-100 นำเลข 9 ตัวนั้นมาทำเป็น matrix 3x3 และ transport และหาเลขอยู่ตำแหน่งที่ (2,2)

```python
list20 = []
for i in range(9):
    number = int(input("กรุณาป้อนตัวเลข"))
    list20.append(number)
arr20 = np.array(list20) 
arr20new = arr20.reshape(3,3)
print(arr20new)
```
