# Quiz 20
แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่ 

```python
def tran_age(x):
    if x <= 10 :
        return '0-10 year'
    elif x <=20 :
        return '11-20 year'
    elif x <=30 :
        return '21-30 year'
    elif x <= 40 :
        return '31-40 year'
    elif x <= 50 :
        return '41-50 year'
    elif x <= 60 :
        return '51-60 year'
    elif x <= 70 :
        return '61-70 year'
    elif x <= 80 :
        return '71-80 year'
```

```python
df['Range_age'] = df['Age'].apply(tran_age)
df
```

```python
servived = pd.crosstab(df.Range_age , df.Survived)
servived
```

```python
servived = (pd.crosstab(df.Range_age , df.Survived , margins = True ,normalize = True)*100).round(2)
servived
```
